--> User gives a filename.


-->Program should show:                                                                                                                                                                     
    Total number of lines                                                                                                                                                                   
    Total words                                                                                                                                                                             
    Total characters                                                                                                                                                                        
    Number of empty lines                                                                                                                                                                   
    Word that appears most                                                                                                                                                                  

--> Skills: file handling, string processing, dictionaries, Looping Logic, Error Handling, Input Handling

```python
#  User gives a filename.

file_name = input("Enter file name: ")

file_ok = True


# --> Total lines
try:
    with open(file_name,"r") as f:
        lines = f.readlines()

except FileNotFoundError:
        print("Error: File not found..")
        file_ok = False

except Exception as E:
        print("Error",E)
        file_ok = False
        
if not file_ok:
    print("Program stopped due to file error.")
else:
        count_line = len(lines)
        print(f"there are {count_line} lines in File")


# --> Total words
try:
    with open(file_name,"r") as f:
        data_word = f.read()

except FileNotFoundError:
        print("Error: File not found..")
        file_ok = False

except Exception as E:
        print("Error",E)
        file_ok = False

if not file_ok:
    print("Program stopped due to file error.")
else:
        # count_word = 0
        # word = ""
        # for i in data_word + " ":
        #     if i not in (" ","\n","\t"):
        #         word += i
        #     else:
        #         if word != "":
        #             count_word += 1
        #             word = ""
                        
        """                     or                      """
        count_word = len(data_word.split())
        print(f"There are {count_word} words in file")


# --> Total characters(excluding "\n")
try:
    with open(file_name,"r") as f:
        data_ch = f.read()

except FileNotFoundError:
        print("Error: File not found..")
        file_ok = False
except Exception as E:
        print("Error",E)
        file_ok = False
        
if not file_ok:
    print("Program stopped due to file error.")
else:
        count_ch = 0
        for i in data_ch:
                if i not in ("\n"):
                        count_ch += 1

        print(f"There are {count_ch} characters in file")


# --> Number of empty lines
try: 
    with open(file_name,"r") as f:
        data_empty = f.readlines()

except FileNotFoundError:
        print("Error: File not found..")
        file_ok = False

except Exception as E:
        print("Error",E)
        file_ok = False

if not file_ok:
    print("Program stopped due to file error.")
else:
        count_empty_line = 0
        for i in range(len(data_empty)):
                if data_empty[i] == "\n":
                        count_empty_line += 1
            
        print(f"There are {count_empty_line} lines empty in file")


#  -->Word that appears most
try:
    with open(file_name,"r") as f:
        temp = f.read()

except FileNotFoundError:
        print("Error: File not found..")
        file_ok = False

except Exception as E:
        print("Error",E)
        file_ok = False
    
if not file_ok:
    print("Program stopped due to file error.")
else:
        data_empty = temp.split()
        data_dict = {}
    
        for i in data_empty:
                if not i[-1].isalnum():
                        i = i[:-1]

                i = i.lower()

                if i in data_dict:
                        data_dict[i] += 1
                else:
                        data_dict[i] = 1

        print(f"characcters count: {data_dict}")
        print(f"most time characters is '{max(data_dict,key = data_dict.get)}' and it comes {max(data_dict.values())} times")
```
