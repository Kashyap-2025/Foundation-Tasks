--> Read both files (with exception handling)

```python
import json

file_ok = True

try:
    
    with open("3_employees1.json") as f:
        emp1_json = json.load(f)

except  json.JSONDecodeError:
    print("Invalid JSON")
    file_ok = False

except FileNotFoundError:
        print("Error: File not found..")
        file_ok = False

except Exception as E:
        print("Error",E)
        file_ok = False

if not file_ok:
    print("Program stopped due to file error")
else:
    #  print(emp1_json)
    print("successfully reading first file")


try:
    
    with open("3_employees2.json") as f:
        emp2_json = json.load(f)

except  json.JSONDecodeError:
    print("Invalid JSON")
    file_ok = False

except FileNotFoundError:
        print("Error: File not found..")
        file_ok = False

except Exception as E:
        print("Error",E)
        file_ok = False

if not file_ok:
    print("Program stopped due to file error")
else:
    #  print(emp2_json)
    print("successfully reading second file")
```

--> Extract the lists from both files

```python
list_emp1 = emp1_json["employees"]
list_emp2 = emp2_json["employees"]

print(list_emp1)
print(list_emp2)

print(len(list_emp1))
print(len(list_emp2))
```

--> Merge the two lists

```python
merge_list = list_emp1 + list_emp2

# print(merge_list)
# print(len(merge_list))

print("successfully merge two lists")
```

--> Remove duplicates

```python
clear_data = {}
duplicate_data = {}
for i in merge_list:
    if i["id"] not in clear_data:
        clear_data.update({i["id"]:i})
    else:
        duplicate_data.update({i["id"]:i})

print(f"There are {len(clear_data)} unique data")
print(f"There are {len(duplicate_data)} duplicate data")

print("successfully remove duplicates")
```

--> Convert dictionary back to list

```python
clear_list = clear_data.values()
# print(clear_list)
print("successfull converted dictionary bake to list")
```

--> Sort the result

```python
# data is allready sorted by id

clear_dict = {}

clear_dict.update({"employees" : list(clear_list)})

# print(clear_dict)
print("successfull sorted  the result")
```

--> Save to a new JSON file

```python
import json
try:
    with open("3_cleardata.json","w") as f:
        json.dump(clear_dict,f,indent=4)
except:
    print("file is not made...")
else:
    print("Succesfully save to a new JSON file")
```
