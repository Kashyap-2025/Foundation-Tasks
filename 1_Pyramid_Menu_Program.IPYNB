--> Pyramid Menu Program

--> creating a menu-driven program where the user can select:

    1.howmany line you want to show  | 2.What type of pyramid they want to print

--> 4 type of pyramid:
--Left-aligned pyramid
--Right-aligned pyramid
--Full pyramid
--Inverted pyramid


--> In this task i Learned and pratice These skills:                                                                                                                                        
1. Menu-Driven Program Design | 2. Pattern Logic | 3. Conditional Logic | 4. Mathematical Pattern Thinking | 5. Functions (optional but recommended) | 6. Clean Code Structure | 
    7. User Interaction

```python
while True:
    print("""*****Pyramid Menu Program*****
          
----------
1 → Left-aligned pyramid
2 → Right-aligned pyramid
3 → Full pyramid
4 → Inverted pyramid
5 → Exit
----------""")
    
    type = abs(int(input("which type of pyramid want to show: ")))
    
    if type == 5:
        print("See you...")
        break

    elif type > 5:
        print("Please enter valid number...")

    else:
        line = abs(int(input("Enter howmany line you want to show: ")))

        if type == 1:
            for i in range(1,line + 1):
                for j in range(1,i+1):
                    print("*",end=" ")
                print()
            print()

        elif type == 2:
            for i in range(line,0,-1):
                for j in range(1,i+1):
                    print("*",end=" ")
                print() 
            print()       

        elif type == 3:
            for i in range(1,line + 1):
                for j in range(line - i):
                    print(" ",end=" ")
                for k in range(2 * i - 1):
                    print("*",end=" ")
                print()
            print()

        elif type == 4:
            for i in range(line,0,-1):
                for j in range(line - i):
                    print(" ",end=" ")
                for k in range(2*i - 1):
                    print("*",end=" ")
                print()
            print()
```
